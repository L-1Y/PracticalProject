```python
import requests
import re
import csv
```

```python
def getpage(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36 Edg/90.0.818.46'}
    response = requests.get(url,headers=headers)
    
    if response.status_code == 200:
        return response.text
    return None
```

```python
url = 'https://search.jd.com/Search?keyword=%E7%BC%96%E7%A8%8B%E4%B9%A6%E7%B1%8D&wq=%E7%BC%96%E7%A8%8B%E4%B9%A6%E7%B1%8D&psort=3&click=0'
html = getpage(url)
print(html)
```

```python
bookname=re.findall('<div class="p-name">.*?<em>.*?</font>(.*?)</em>',html,re.S)
print(bookname)
```

```python
price=re.findall('<div class="p-price">.*?<em>(.*?)</em><i>(.*?)</i>',html,re.S)
print(price)
```

```python
zd1=list()
for j in range (0,10):
    zd = dict() 
    zd["name"]=bookname[j]
    zd["jiage"]=price[j][-1]
    zd1.append(zd)
```

```python
with open('jingdong.csv','w',newline='',encoding='utf-8-sig')as f :
    writer=csv.DictWriter(f,fieldnames=['name','jiage'])
    writer.writeheader()
    writer.writerows(zd1)
print(zd1)
```
